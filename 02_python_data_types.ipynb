## Overview
Python has several built-in data types that are used to define the operations possible on them and the storage method for each of them. Understanding data types is fundamental to Python programming.

## Built-in Data Types

### 1. Numeric Types
- **int**: Integer numbers (whole numbers)
- **float**: Floating-point numbers (decimal numbers)
- **complex**: Complex numbers

### 2. Sequence Types
- **str**: String (text)
- **list**: Mutable sequence
- **tuple**: Immutable sequence

### 3. Mapping Type
- **dict**: Dictionary (key-value pairs)

### 4. Set Types
- **set**: Mutable set
- **frozenset**: Immutable set

### 5. Boolean Type
- **bool**: True or False

### 6. None Type
- **None**: Represents absence of value

## Numeric Data Types

### Integer (int)
Whole numbers, positive or negative, without decimal points.

```python
# Integer examples
positive_int = 42
negative_int = -17
zero = 0
large_int = 1234567890


print(f"Positive integer: {positive_int}, type: {type(positive_int)}")
print(f"Negative integer: {negative_int}, type: {type(negative_int)}")
print(f"Zero: {zero}, type: {type(zero)}")
print(f"Large integer: {large_int}, type: {type(large_int)}")

# Integer operations
a = 10
b = 3
print(f"Addition: {a} + {b} = {a + b}")
print(f"Subtraction: {a} - {b} = {a - b}")
print(f"Multiplication: {a} * {b} = {a * b}")
print(f"Division: {a} / {b} = {a / b} (type: {type(a / b)})")
print(f"Floor division: {a} // {b} = {a // b} (type: {type(a // b)})")
print(f"Modulus: {a} % {b} = {a % b}")
print(f"Power: {a} ** {b} = {a ** b}")
```

### Float (float)
Decimal numbers with floating-point precision.

```python
# Float examples
pi = 3.14159
e = 2.71828
negative_float = -0.001
scientific = 1.23e-4  # 1.23 × 10^-4

print(f"Pi: {pi}, type: {type(pi)}")
print(f"Euler's number: {e}, type: {type(e)}")
print(f"Negative float: {negative_float}, type: {type(negative_float)}")
print(f"Scientific notation: {scientific}, type: {type(scientific)}")

# Float operations
x = 3.5
y = 2.1
print(f"x = {x}, y = {y}")
print(f"Addition: {x} + {y} = {x + y}")
print(f"Multiplication: {x} * {y} = {x * y}")
print(f"Division: {x} / {y} = {x / y}")

# Converting between types
int_to_float = float(42)
float_to_int = int(3.9)  # Truncates decimal part
print(f"42 to float: {int_to_float}, type: {type(int_to_float)}")
print(f"3.9 to int: {float_to_int}, type: {type(float_to_int)}")
```

### Complex (complex)
Numbers with real and imaginary parts.

```python
# Complex number examples
c1 = 3 + 4j
c2 = complex(1, 2)  # 1 + 2j
c3 = 5j  # 0 + 5j

print(f"Complex 1: {c1}, type: {type(c1)}")
print(f"Complex 2: {c2}, type: {type(c2)}")
print(f"Complex 3: {c3}, type: {type(c3)}")

# Accessing real and imaginary parts
print(f"Real part of {c1}: {c1.real}")
print(f"Imaginary part of {c1}: {c1.imag}")

# Complex arithmetic
print(f"{c1} + {c2} = {c1 + c2}")
print(f"{c1} * {c2} = {c1 * c2}")
print(f"Conjugate of {c1}: {c1.conjugate()}")
print(f"Absolute value of {c1}: {abs(c1)}")
```

## String Data Type (str)
Immutable sequence of characters.

```python
# String creation
single_quotes = 'Hello, World!'
double_quotes = "Python Programming"
triple_quotes = """
Multi-line
string
with triple quotes
"""

print(f"Single quotes: {single_quotes}")
print(f"Double quotes: {double_quotes}")
print(f"Triple quotes: {triple_quotes}")
print(f"Type: {type(single_quotes)}")

# String indexing and slicing
text = "Python Programming"
print(f"Original text: {text}")
print(f"Length: {len(text)}")
print(f"First character: {text[0]}")
print(f"Last character: {text[-1]}")
print(f"First 6 characters: {text[:6]}")
print(f"Last 11 characters: {text[-11:]}")
print(f"Characters 7-11: {text[7:12]}")
print(f"Every second character: {text[::2]}")
print(f"Reverse: {text[::-1]}")
```

```python
# String methods
sample_text = "  Python Programming  "

print(f"Original: '{sample_text}'")
print(f"Upper: '{sample_text.upper()}'")
print(f"Lower: '{sample_text.lower()}'")
print(f"Title: '{sample_text.title()}'")
print(f"Capitalize: '{sample_text.capitalize()}'")
print(f"Stripped: '{sample_text.strip()}'")
print(f"Left stripped: '{sample_text.lstrip()}'")
print(f"Right stripped: '{sample_text.rstrip()}'")

# String splitting and joining
words = "Python,Java,C++,JavaScript".split(',')
print(f"Split result: {words}")
joined = " - ".join(words)
print(f"Joined result: {joined}")

# String replacement
text = "I love Python programming"
replaced = text.replace("Python", "Java")
print(f"Original: {text}")
print(f"Replaced: {replaced}")

# String formatting
name = "Alice"
age = 25
formatted = "{} is {} years old".format(name, age)
f_string = f"{name} is {age} years old"
print(f"Format method: {formatted}")
print(f"F-string: {f_string}")
```

## List Data Type (list)
Mutable, ordered sequence of elements.

```python
# List creation
empty_list = []
numbers = [1, 2, 3, 4, 5]
mixed = [1, "hello", 3.14, True, [1, 2, 3]]
from_range = list(range(5))

print(f"Empty list: {empty_list}, type: {type(empty_list)}")
print(f"Numbers: {numbers}, type: {type(numbers)}")
print(f"Mixed: {mixed}, type: {type(mixed)}")
print(f"From range: {from_range}, type: {type(from_range)}")

# List indexing and slicing
fruits = ["apple", "banana", "cherry", "date", "elderberry"]
print(f"Original list: {fruits}")
print(f"Length: {len(fruits)}")
print(f"First element: {fruits[0]}")
print(f"Last element: {fruits[-1]}")
print(f"First 3 elements: {fruits[:3]}")
print(f"Last 3 elements: {fruits[-3:]}")
print(f"Elements 1-3: {fruits[1:4]}")
print(f"Every second element: {fruits[::2]}")
print(f"Reverse: {fruits[::-1]}")
```

```python
# List methods and operations
numbers = [3, 1, 4, 1, 5, 9, 2, 6]

print(f"Original: {numbers}")

# Adding elements
numbers.append(8)
print(f"After append(8): {numbers}")

numbers.insert(0, 0)
print(f"After insert(0, 0): {numbers}")

numbers.extend([7, 8, 9])
print(f"After extend([7, 8, 9]): {numbers}")

# Removing elements
removed = numbers.pop()
print(f"Popped element: {removed}")
print(f"After pop(): {numbers}")

numbers.remove(1)
print(f"After remove(1): {numbers}")

# Sorting and reversing
numbers.sort()
print(f"After sort(): {numbers}")

numbers.reverse()
print(f"After reverse(): {numbers}")

# List comprehension
squares = [x**2 for x in range(10)]
even_squares = [x**2 for x in range(10) if x % 2 == 0]
print(f"Squares: {squares}")
print(f"Even squares: {even_squares}")
```

## Tuple Data Type (tuple)
Immutable, ordered sequence of elements.

```python
# Tuple creation
empty_tuple = ()
single_element = (42,)  # Note the comma
coordinates = (3, 4)
person = ("John", 30, "Engineer")
nested = (1, (2, 3), [4, 5])

print(f"Empty tuple: {empty_tuple}, type: {type(empty_tuple)}")
print(f"Single element: {single_element}, type: {type(single_element)}")
print(f"Coordinates: {coordinates}, type: {type(coordinates)}")
print(f"Person: {person}, type: {type(person)}")
print(f"Nested: {nested}, type: {type(nested)}")

# Tuple indexing and slicing
rgb = (255, 128, 0)
print(f"RGB tuple: {rgb}")
print(f"Red component: {rgb[0]}")
print(f"Green component: {rgb[1]}")
print(f"Blue component: {rgb[2]}")
print(f"First two: {rgb[:2]}")
print(f"Last two: {rgb[-2:]}")

# Tuple unpacking
x, y = coordinates
print(f"Unpacked: x = {x}, y = {y}")

name, age, profession = person
print(f"Name: {name}, Age: {age}, Profession: {profession}")

# Tuple methods
numbers = (1, 2, 2, 3, 2, 4, 5)
print(f"Numbers: {numbers}")
print(f"Count of 2: {numbers.count(2)}")
print(f"Index of 3: {numbers.index(3)}")
print(f"Length: {len(numbers)}")
```

## Dictionary Data Type (dict)
Mutable collection of key-value pairs.

```python
# Dictionary creation
empty_dict = {}
person = {"name": "Alice", "age": 30, "city": "New York"}
grades = dict(John=85, Alice=92, Bob=78)
nested = {"user": {"name": "Bob", "preferences": ["python", "data"]}}

print(f"Empty dict: {empty_dict}, type: {type(empty_dict)}")
print(f"Person: {person}, type: {type(person)}")
print(f"Grades: {grades}, type: {type(grades)}")
print(f"Nested: {nested}, type: {type(nested)}")

# Accessing dictionary elements
print(f"Name: {person['name']}")
print(f"Age: {person['age']}")
print(f"City: {person['city']}")

# Using get() method (safer)
print(f"Name: {person.get('name')}")
print(f"Phone: {person.get('phone', 'Not provided')}")

# Dictionary methods
print(f"Keys: {list(person.keys())}")
print(f"Values: {list(person.values())}")
print(f"Items: {list(person.items())}")
print(f"Length: {len(person)}")
```

```python
# Dictionary operations
student = {"name": "Charlie", "age": 20}
print(f"Original: {student}")

# Adding/updating elements
student["grade"] = "A"
print(f"After adding grade: {student}")

student.update({"major": "Computer Science", "year": 2})
print(f"After update: {student}")

# Removing elements
removed_grade = student.pop("grade")
print(f"Removed grade: {removed_grade}")
print(f"After pop: {student}")

del student["year"]
print(f"After del: {student}")

# Dictionary comprehension
squares = {x: x**2 for x in range(5)}
print(f"Squares dict: {squares}")

filtered = {k: v for k, v in person.items() if isinstance(v, str)}
print(f"String values only: {filtered}")
```

## Set Data Types

### Set (set)
Mutable, unordered collection of unique elements.

```python
# Set creation
empty_set = set()
numbers = {1, 2, 3, 4, 5}
from_list = set([1, 2, 2, 3, 3, 4])  # Duplicates removed
from_string = set("hello")  # Characters from string

print(f"Empty set: {empty_set}, type: {type(empty_set)}")
print(f"Numbers: {numbers}, type: {type(numbers)}")
print(f"From list: {from_list}, type: {type(from_list)}")
print(f"From string: {from_string}, type: {type(from_string)}")

# Set operations
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

print(f"Set 1: {set1}")
print(f"Set 2: {set2}")
print(f"Union: {set1 | set2}")
print(f"Intersection: {set1 & set2}")
print(f"Difference: {set1 - set2}")
print(f"Symmetric difference: {set1 ^ set2}")

# Set methods
fruits = {"apple", "banana", "cherry"}
print(f"Original: {fruits}")

fruits.add("orange")
print(f"After add: {fruits}")

fruits.remove("banana")
print(f"After remove: {fruits}")

fruits.discard("grape")  # No error if element doesn't exist
print(f"After discard: {fruits}")

popped = fruits.pop()
print(f"Popped: {popped}")
print(f"After pop: {fruits}")
```

### Frozenset (frozenset)
Immutable version of set.

```python
# Frozenset creation
frozen = frozenset([1, 2, 3, 4, 5])
print(f"Frozenset: {frozen}, type: {type(frozen)}")

# Frozenset operations (read-only)
print(f"Length: {len(frozen)}")
print(f"Contains 3: {3 in frozen}")
print(f"Contains 10: {10 in frozen}")

# Set operations with frozenset
set1 = {1, 2, 3}
frozen2 = frozenset([3, 4, 5])

print(f"Set: {set1}")
print(f"Frozenset: {frozen2}")
print(f"Union: {set1 | frozen2}")
print(f"Intersection: {set1 & frozen2}")

# Note: frozenset is immutable
try:
    frozen.add(6)
except AttributeError as e:
    print(f"Error: {e}")
```

## Boolean Data Type (bool)
Represents True or False values.

```python
# Boolean values
true_value = True
false_value = False

print(f"True: {true_value}, type: {type(true_value)}")
print(f"False: {false_value}, type: {type(false_value)}")

# Boolean operations
print(f"True and True: {True and True}")
print(f"True and False: {True and False}")
print(f"True or False: {True or False}")
print(f"False or False: {False or False}")
print(f"not True: {not True}")
print(f"not False: {not False}")

# Truthiness in Python
print(f"bool(0): {bool(0)}")
print(f"bool(1): {bool(1)}")
print(f"bool(''): {bool('')}")
print(f"bool('hello'): {bool('hello')}")
print(f"bool([]): {bool([])}")
print(f"bool([1, 2, 3]): {bool([1, 2, 3])}")
print(f"bool(None): {bool(None)}")
print(f"bool(False): {bool(False)}")
```

## None Data Type
Represents the absence of a value or a null value.

```python
# None examples
none_value = None
print(f"None value: {none_value}, type: {type(none_value)}")

# Common use cases
def find_user(user_id):
    # Simulate database lookup
    users = {1: "Alice", 2: "Bob", 3: "Charlie"}
    return users.get(user_id, None)

result1 = find_user(2)
result2 = find_user(99)

print(f"User 2: {result1}")
print(f"User 99: {result2}")
print(f"Is result2 None? {result2 is None}")

# None in comparisons
print(f"None == None: {None == None}")
print(f"None is None: {None is None}")
print(f"None != 0: {None != 0}")
print(f"None != '': {None != ''}")
```

## Type Checking and Conversion

### Type Checking
- `type()` function
- `isinstance()` function

### Type Conversion
- `int()`, `float()`, `str()`, `list()`, `tuple()`, `set()`, `dict()`

```python
# Type checking
value = 42
print(f"Value: {value}")
print(f"Type: {type(value)}")
print(f"Is int: {isinstance(value, int)}")
print(f"Is float: {isinstance(value, float)}")
print(f"Is number: {isinstance(value, (int, float))}")

# Type conversion examples
print("\nType Conversions:")
print(f"int(3.14): {int(3.14)}")
print(f"float(42): {float(42)}")
print(f"str(123): {str(123)}, type: {type(str(123))}")
print(f"list('hello'): {list('hello')}")
print(f"tuple([1, 2, 3]): {tuple([1, 2, 3])}")
print(f"set([1, 2, 2, 3]): {set([1, 2, 2, 3])}")
print(f"dict([('a', 1), ('b', 2)]): {dict([('a', 1), ('b', 2)])}")

# Safe type conversion
def safe_int(value):
    try:
        return int(value)
    except (ValueError, TypeError):
        return None

print(f"\nSafe conversions:")
print(f"safe_int('42'): {safe_int('42')}")
print(f"safe_int('abc'): {safe_int('abc')}")
print(f"safe_int(3.14): {safe_int(3.14)}")
```
