## Overview
Conditional statements allow your program to make decisions and execute different code blocks based on whether certain conditions are true or false. Python provides several ways to implement conditional logic.

## Types of Conditional Statements

### 1. If Statement
The basic conditional statement that executes code only when a condition is True.

### 2. If-Else Statement
Provides an alternative path when the condition is False.

### 3. If-Elif-Else Statement
Handles multiple conditions with multiple code paths.

### 4. Nested If Statements
If statements inside other if statements for complex logic.

### 5. Match Statement (Python 3.10+)
Pattern matching for complex conditional logic.

### 6. Conditional Expressions (Ternary Operator)
Compact one-line conditional statements.

## Basic If Statement
The simplest form of conditional execution.

```python
# Basic if statement
age = 18

if age >= 18:
    print("You are an adult")
    print("You can vote")
    print("You can drive")

print("This statement always executes")
```

```python
# If statement with different conditions
temperature = 25

if temperature > 30:
    print("It's hot today!")

if temperature < 10:
    print("It's cold today!")

if temperature >= 20 and temperature <= 30:
    print("It's pleasant weather!")

print("Weather check completed")
```

```python
# If statement with string conditions
username = "admin"
password = "12345"

if username == "admin":
    print("Username is correct")
    
    if password == "12345":
        print("Password is correct")
        print("Access granted!")
    else:
        print("Password is incorrect")
        print("Access denied!")
else:
    print("Username is incorrect")
    print("Access denied!")
```

## If-Else Statement
Provides an alternative path when the condition is False.

```python
# Basic if-else statement
score = 75

if score >= 60:
    print("Pass")
    print("Congratulations!")
else:
    print("Fail")
    print("Better luck next time")

print("Grade evaluation completed")
```

```python
# If-else with different data types
value = 42

if isinstance(value, int):
    if value > 0:
        print(f"{value} is a positive integer")
    else:
        print(f"{value} is a non-positive integer")
elif isinstance(value, float):
    print(f"{value} is a float")
elif isinstance(value, str):
    print(f"'{value}' is a string")
else:
    print(f"{value} is of type {type(value)}")

print("Type check completed")
```

```python
# If-else with list operations
numbers = [1, 2, 3, 4, 5]

if len(numbers) > 0:
    if len(numbers) == 1:
        print(f"List has one element: {numbers[0]}")
    else:
        print(f"List has {len(numbers)} elements")
        print(f"First element: {numbers[0]}")
        print(f"Last element: {numbers[-1]}")
        
        if len(numbers) % 2 == 0:
            print("List has even number of elements")
        else:
            print("List has odd number of elements")
else:
    print("List is empty")

print("List analysis completed")
```

## If-Elif-Else Statement
Handles multiple conditions with multiple code paths.

```python
# Basic if-elif-else statement
grade = 85

if grade >= 90:
    print("Grade: A")
    print("Excellent!")
elif grade >= 80:
    print("Grade: B")
    print("Good job!")
elif grade >= 70:
    print("Grade: C")
    print("Satisfactory")
elif grade >= 60:
    print("Grade: D")
    print("Needs improvement")
else:
    print("Grade: F")
    print("Failed")

print("Grade evaluation completed")
```

```python
# If-elif-else with string matching
day = "Monday"

if day == "Monday":
    print("Start of the work week")
    print("Time to get back to work")
elif day == "Tuesday":
    print("Second day of the week")
elif day == "Wednesday":
    print("Middle of the week")
    print("Hump day!")
elif day == "Thursday":
    print("Almost Friday")
elif day == "Friday":
    print("TGIF! Weekend is coming")
elif day in ["Saturday", "Sunday"]:
    print("Weekend! Time to relax")
else:
    print("Invalid day")

print("Day analysis completed")
```

```python
# If-elif-else with numerical ranges
age = 25
income = 50000

if age < 18:
    category = "Minor"
    loan_eligible = False
elif age < 25:
    category = "Young Adult"
    if income >= 30000:
        loan_eligible = True
    else:
        loan_eligible = False
elif age < 65:
    category = "Working Age"
    if income >= 25000:
        loan_eligible = True
    else:
        loan_eligible = False
else:
    category = "Senior"
    if income >= 20000:
        loan_eligible = True
    else:
        loan_eligible = False

print(f"Age: {age}, Category: {category}")
print(f"Income: ${income:,}")
print(f"Loan Eligible: {loan_eligible}")
```

## Nested If Statements
If statements inside other if statements for complex logic.

```python
# Nested if statements
age = 25
income = 50000
credit_score = 750

print("Loan Application Evaluation")
print("=" * 30)

if age >= 18:
    print("✓ Age requirement met")
    
    if income >= 30000:
        print("✓ Income requirement met")
        
        if credit_score >= 700:
            print("✓ Credit score requirement met")
            print("🎉 LOAN APPROVED!")
            
            # Determine loan amount based on income
            if income >= 50000:
                loan_amount = 100000
            elif income >= 40000:
                loan_amount = 75000
            else:
                loan_amount = 50000
            
            print(f"Approved loan amount: ${loan_amount:,}")
            
        else:
            print(f"✗ Credit score too low: {credit_score} (minimum: 700)")
            print("Loan denied due to poor credit score")
    else:
        print(f"✗ Income too low: ${income:,} (minimum: $30,000)")
        print("Loan denied due to insufficient income")
else:
    print(f"✗ Age too young: {age} (minimum: 18)")
    print("Loan denied due to age requirement")

print("\nEvaluation completed")
```

```python
# Nested if with complex logic
def evaluate_student(score, attendance, participation):
    """Evaluate student performance with nested conditions"""
    print(f"Student Evaluation - Score: {score}, Attendance: {attendance}%, Participation: {participation}/10")
    print("=" * 60)
    
    if score >= 90:
        grade = "A"
        if attendance >= 90:
            if participation >= 8:
                status = "Outstanding Student"
                scholarship = True
            else:
                status = "Excellent Student"
                scholarship = False
        else:
            status = "Good Student (attendance issue)"
            scholarship = False
    
    elif score >= 80:
        grade = "B"
        if attendance >= 85:
            status = "Good Student"
        else:
            status = "Good Student (attendance concern)"
        scholarship = False
    
    elif score >= 70:
        grade = "C"
        if attendance >= 80:
            status = "Satisfactory Student"
        else:
            status = "Student needs improvement"
        scholarship = False
    
    elif score >= 60:
        grade = "D"
        status = "Student at risk"
        scholarship = False
    
    else:
        grade = "F"
        status = "Student failed"
        scholarship = False
    
    print(f"Grade: {grade}")
    print(f"Status: {status}")
    print(f"Scholarship Eligible: {scholarship}")
    
    return grade, status, scholarship

# Test the function
print("Test Case 1:")
result1 = evaluate_student(95, 95, 9)

print("\nTest Case 2:")
result2 = evaluate_student(85, 80, 7)

print("\nTest Case 3:")
result3 = evaluate_student(55, 70, 5)
```

## Match Statement (Python 3.10+)
Pattern matching for complex conditional logic.

```python
# Match statement with simple patterns
def analyze_command(command):
    """Analyze command using match statement"""
    match command.split():
        case ["help"]:
            return "Showing help information"
        case ["search", query]:
            return f"Searching for: {query}"
        case ["download", filename]:
            return f"Downloading: {filename}"
        case ["download", filename, "--force"]:
            return f"Force downloading: {filename}"
        case ["quit"] | ["exit"]:
            return "Goodbye!"
        case _:
            return "Unknown command"

# Test the match statement
commands = ["help", "search python", "download file.txt", "download file.txt --force", "quit"]

print("Command Analysis with Match Statement")
print("=" * 40)
for cmd in commands:
    result = analyze_command(cmd)
    print(f"Command: '{cmd}' -> {result}")
```

```python
# Match statement with complex patterns
def process_data(data):
    """Process different types of data using match statement"""
    match data:
        case None:
            return "No data provided"
        case str() as text if len(text) > 100:
            return f"Long text: {len(text)} characters"
        case str() as text:
            return f"Short text: {text}"
        case int() as num if num < 0:
            return f"Negative integer: {num}"
        case int() as num if num == 0:
            return "Zero"
        case int() as num:
            return f"Positive integer: {num}"
        case float() as num:
            return f"Float: {num}"
        case list() as items if len(items) == 0:
            return "Empty list"
        case list() as items:
            return f"List with {len(items)} items: {items}"
        case dict() as mapping if len(mapping) == 0:
            return "Empty dictionary"
        case dict() as mapping:
            return f"Dictionary with {len(mapping)} keys: {list(mapping.keys())}"
        case _:
            return f"Unknown data type: {type(data)}"

# Test the function
test_data = [
    None,
    "Short text",
    "This is a very long text that exceeds one hundred characters to test the long text condition in our match statement",
    -5,
    0,
    42,
    3.14,
    [],
    [1, 2, 3],
    {},
    {"name": "Alice", "age": 30}
]

print("Data Processing with Match Statement")
print("=" * 40)
for item in test_data:
    result = process_data(item)
    print(f"{item} -> {result}")
```

## Conditional Expressions (Ternary Operator)
Compact one-line conditional statements.

```python
# Basic conditional expression
age = 20
status = "adult" if age >= 18 else "minor"
print(f"Age {age}: {status}")

# Conditional expression with different types
score = 85
grade = "A" if score >= 90 else "B" if score >= 80 else "C" if score >= 70 else "D" if score >= 60 else "F"
print(f"Score {score}: Grade {grade}")

# Conditional expression in function calls
def greet(name, is_formal=True):
    greeting = "Good day, " if is_formal else "Hey, "
    return greeting + name

print(greet("Alice", True))
print(greet("Bob", False))

# Conditional expression with lists
numbers = [1, 2, 3, 4, 5]
result = "Even count" if len([x for x in numbers if x % 2 == 0]) > len([x for x in numbers if x % 2 != 0]) else "Odd count"
print(f"Numbers {numbers}: {result}")
```

```python
# Complex conditional expressions
def analyze_number(num):
    """Analyze number using conditional expressions"""
    category = "positive" if num > 0 else "negative" if num < 0 else "zero"
    size = "large" if abs(num) > 100 else "medium" if abs(num) > 10 else "small"
    parity = "even" if num % 2 == 0 else "odd"
    
    return f"{num} is a {size} {category} {parity} number"

# Test the function
test_numbers = [-150, -25, -1, 0, 1, 25, 150]

print("Number Analysis with Conditional Expressions")
print("=" * 45)
for num in test_numbers:
    result = analyze_number(num)
    print(result)
```

## Logical Operators in Conditionals
Combining multiple conditions using logical operators.

```python
# Logical AND operator
age = 25
income = 50000
credit_score = 750

if age >= 18 and income >= 30000 and credit_score >= 700:
    print("All requirements met - Loan approved!")
else:
    print("Some requirements not met - Loan denied")
    
    if age < 18:
        print("  - Age requirement not met")
    if income < 30000:
        print(f"  - Income requirement not met (${income:,} < $30,000)")
    if credit_score < 700:
        print(f"  - Credit score requirement not met ({credit_score} < 700)")
```

```python
# Logical OR operator
payment_method = "credit_card"
amount = 150

if payment_method == "cash" or payment_method == "debit_card" or amount < 100:
    discount = 0.05  # 5% discount
    print(f"Eligible for {discount*100}% discount")
else:
    discount = 0.0
    print("No discount available")

final_amount = amount * (1 - discount)
print(f"Original amount: ${amount}")
print(f"Discount: ${amount * discount:.2f}")
print(f"Final amount: ${final_amount:.2f}")
```

```python
# Logical NOT operator
is_weekend = False
is_holiday = False
is_vacation = True

if not (is_weekend or is_holiday or is_vacation):
    print("It's a regular work day")
    print("Time to go to work")
else:
    print("It's not a regular work day")
    
    if is_weekend:
        print("  - It's the weekend!")
    if is_holiday:
        print("  - It's a holiday!")
    if is_vacation:
        print("  - You're on vacation!")
    
    print("Enjoy your time off!")
```

## Practical Examples

### Example 1: Grade Calculator with Multiple Criteria

```python
def calculate_grade(assignments, midterm, final, participation):
    """Calculate final grade based on multiple criteria"""
    print("Grade Calculation")
    print("=" * 20)
    
    # Validate inputs
    if not (0 <= assignments <= 100 and 0 <= midterm <= 100 and 
            0 <= final <= 100 and 0 <= participation <= 10):
        return "Invalid input: Scores must be within valid ranges"
    
    # Calculate weighted grade
    weighted_grade = (assignments * 0.3 + midterm * 0.3 + 
                     final * 0.35 + participation * 5)
    
    # Determine letter grade
    if weighted_grade >= 93:
        letter_grade = "A"
        quality_points = 4.0
    elif weighted_grade >= 90:
        letter_grade = "A-"
        quality_points = 3.7
    elif weighted_grade >= 87:
        letter_grade = "B+"
        quality_points = 3.3
    elif weighted_grade >= 83:
        letter_grade = "B"
        quality_points = 3.0
    elif weighted_grade >= 80:
        letter_grade = "B-"
        quality_points = 2.7
    elif weighted_grade >= 77:
        letter_grade = "C+"
        quality_points = 2.3
    elif weighted_grade >= 73:
        letter_grade = "C"
        quality_points = 2.0
    elif weighted_grade >= 70:
        letter_grade = "C-"
        quality_points = 1.7
    elif weighted_grade >= 67:
        letter_grade = "D+"
        quality_points = 1.3
    elif weighted_grade >= 63:
        letter_grade = "D"
        quality_points = 1.0
    elif weighted_grade >= 60:
        letter_grade = "D-"
        quality_points = 0.7
    else:
        letter_grade = "F"
        quality_points = 0.0
    
    # Determine status
    if weighted_grade >= 60:
        status = "PASS"
        if weighted_grade >= 90:
            status += " (Honors)"
    else:
        status = "FAIL"
    
    # Print results
    print(f"Assignments: {assignments}%")
    print(f"Midterm: {midterm}%")
    print(f"Final: {final}%")
    print(f"Participation: {participation}/10")
    print(f"\nWeighted Grade: {weighted_grade:.1f}%")
    print(f"Letter Grade: {letter_grade}")
    print(f"Quality Points: {quality_points}")
    print(f"Status: {status}")
    
    return letter_grade, weighted_grade, status

# Test the function
print("Test Case 1: Excellent Student")
result1 = calculate_grade(95, 92, 94, 10)

print("\n" + "="*50 + "\n")

print("Test Case 2: Average Student")
result2 = calculate_grade(75, 78, 80, 7)

print("\n" + "="*50 + "\n")

print("Test Case 3: Failing Student")
result3 = calculate_grade(45, 50, 55, 3)
```

### Example 2: User Authentication System

```python
def authenticate_user(username, password, user_database):
    """Authenticate user with multiple security checks"""
    print(f"Authentication attempt for user: {username}")
    print("=" * 50)
    
    # Check if username exists
    if username not in user_database:
        print("✗ Authentication failed: Username not found")
        return False, "Invalid username or password"
    
    user_info = user_database[username]
    
    # Check if account is locked
    if user_info.get('locked', False):
        print("✗ Authentication failed: Account is locked")
        return False, "Account is locked. Contact administrator."
    
    # Check if account is expired
    if 'expiry_date' in user_info:
        from datetime import datetime
        current_date = datetime.now().date()
        if current_date > user_info['expiry_date']:
            print("✗ Authentication failed: Account expired")
            return False, "Account has expired. Please renew."
    
    # Check password
    if user_info['password'] == password:
        print("✓ Password correct")
        
        # Check if password needs to be changed
        if user_info.get('password_expired', False):
            print("⚠ Password expired - must be changed")
            return True, "Authentication successful but password must be changed"
        
        # Check if 2FA is required
        if user_info.get('requires_2fa', False):
            print("⚠ 2FA required for this account")
            return True, "Authentication successful but 2FA required"
        
        # Check user role and permissions
        role = user_info.get('role', 'user')
        if role == 'admin':
            print("✓ Admin access granted")
            return True, "Admin authentication successful"
        elif role == 'moderator':
            print("✓ Moderator access granted")
            return True, "Moderator authentication successful"
        else:
            print("✓ User access granted")
            return True, "User authentication successful"
    else:
        print("✗ Authentication failed: Incorrect password")
        
        # Increment failed attempts
        failed_attempts = user_info.get('failed_attempts', 0) + 1
        user_info['failed_attempts'] = failed_attempts
        
        # Lock account after 3 failed attempts
        if failed_attempts >= 3:
            user_info['locked'] = True
            print(f"⚠ Account locked after {failed_attempts} failed attempts")
            return False, "Account locked due to multiple failed attempts"
        
        return False, f"Invalid password. {3 - failed_attempts} attempts remaining."

# Test database
test_users = {
    'admin': {
        'password': 'admin123',
        'role': 'admin',
        'requires_2fa': True
    },
    'user1': {
        'password': 'password123',
        'role': 'user',
        'failed_attempts': 0
    },
    'locked_user': {
        'password': 'secret',
        'role': 'user',
        'locked': True
    }
}

print("User Authentication System Test")
print("=" * 50)

# Test successful authentication
print("\nTest 1: Successful admin authentication")
success1, message1 = authenticate_user('admin', 'admin123', test_users)

print("\nTest 2: Failed authentication")
success2, message2 = authenticate_user('user1', 'wrongpass', test_users)

print("\nTest 3: Locked account")
success3, message3 = authenticate_user('locked_user', 'secret', test_users)
```
