## Overview
Python syntax is known for its simplicity and readability. It emphasizes code readability with its notable use of significant whitespace and clean, minimal syntax.

## Key Features of Python Syntax

### 1. Indentation
Python uses indentation to define code blocks instead of curly braces or keywords.

### 2. Simple and Readable
Python syntax is designed to be intuitive and easy to read.

### 3. Dynamic Typing
Variables don't need type declarations.

### 4. High-Level Language
Python abstracts away low-level details.

## Basic Syntax Rules

### 1. Comments
- Single line comments use `#`
- Multi-line comments use triple quotes `"""` or `'''`

### 2. Variables and Assignment
- Use `=` for assignment
- Variable names are case-sensitive
- Can contain letters, numbers, and underscores

### 3. Indentation
- Use consistent indentation (typically 4 spaces)
- Indentation defines code blocks

### 4. Line Continuation
- Use `\` for line continuation
- Or use parentheses for implicit continuation

```python
# This is a single line comment

"""
This is a multi-line comment
It can span multiple lines
Useful for documentation
"""

print("Hello, World!")
```

```python
# Variable assignment
name = "Python"
version = 3.9
is_awesome = True

print(f"Language: {name}")
print(f"Version: {version}")
print(f"Awesome: {is_awesome}")
```

```python
# Indentation example
if True:
    print("This is indented")
    print("This is also indented")
    if False:
        print("This is nested indentation")
print("This is not indented")
```

```python
# Line continuation with backslash
long_string = "This is a very long string that " \
               "spans multiple lines using backslash"

print(long_string)

# Implicit line continuation with parentheses
result = (1 + 2 + 3 + 
          4 + 5 + 6)

print(f"Result: {result}")
```

## Naming Conventions

### 1. Variables and Functions
- Use lowercase with underscores (snake_case)
- Example: `user_name`, `calculate_total`

### 2. Constants
- Use uppercase with underscores
- Example: `MAX_VALUE`, `PI`

### 3. Classes
- Use PascalCase (CapitalizedWords)
- Example: `UserAccount`, `DatabaseConnection`

### 4. Private Variables
- Use single underscore prefix
- Example: `_private_var`

```python
# Naming convention examples

# Variables (snake_case)
user_name = "John Doe"
age = 25
is_active = True

# Constants (UPPER_CASE)
MAX_ATTEMPTS = 3
PI = 3.14159
API_BASE_URL = "https://api.example.com"

# Functions (snake_case)
def calculate_area(radius):
    return PI * radius ** 2

def get_user_info(user_id):
    return f"User ID: {user_id}"

# Classes (PascalCase)
class UserAccount:
    def __init__(self, username):
        self.username = username
        self._private_var = "hidden"  # Private variable

print(f"Area: {calculate_area(5):.2f}")
print(f"User info: {get_user_info(123)}")

user = UserAccount("john_doe")
print(f"Username: {user.username}")
```

## Basic Operators

### 1. Arithmetic Operators
- `+` Addition
- `-` Subtraction
- `*` Multiplication
- `/` Division (float)
- `//` Floor division
- `%` Modulus
- `**` Exponentiation

### 2. Comparison Operators
- `==` Equal to
- `!=` Not equal to
- `>` Greater than
- `<` Less than
- `>=` Greater than or equal to
- `<=` Less than or equal to

### 3. Logical Operators
- `and` Logical AND
- `or` Logical OR
- `not` Logical NOT

```python
# Arithmetic operators
a = 10
b = 3

print(f"a = {a}, b = {b}")
print(f"Addition: {a} + {b} = {a + b}")
print(f"Subtraction: {a} - {b} = {a - b}")
print(f"Multiplication: {a} * {b} = {a * b}")
print(f"Division: {a} / {b} = {a / b}")
print(f"Floor Division: {a} // {b} = {a // b}")
print(f"Modulus: {a} % {b} = {a % b}")
print(f"Exponentiation: {a} ** {b} = {a ** b}")
```

```python
# Comparison operators
x = 5
y = 10

print(f"x = {x}, y = {y}")
print(f"x == y: {x == y}")
print(f"x != y: {x != y}")
print(f"x > y: {x > y}")
print(f"x < y: {x < y}")
print(f"x >= y: {x >= y}")
print(f"x <= y: {x <= y}")
```

```python
# Logical operators
p = True
q = False

print(f"p = {p}, q = {q}")
print(f"p and q: {p and q}")
print(f"p or q: {p or q}")
print(f"not p: {not p}")
print(f"not q: {not q}")
print(f"(p and q) or (not p): {(p and q) or (not p)}")
```

## String Syntax

### 1. String Creation
- Single quotes: `'Hello'`
- Double quotes: `"Hello"`
- Triple quotes: `"""Hello"""` or `'''Hello'''`

### 2. String Methods
- `.upper()`, `.lower()`
- `.strip()`, `.split()`
- `.join()`, `.replace()`

### 3. String Formatting
- f-strings (Python 3.6+)
- `.format()` method
- `%` operator (old style)

```python
# String creation
single_quotes = 'Hello, World!'
double_quotes = "Python Programming"
triple_quotes = """
This is a multi-line string
It can contain multiple lines
Very useful for documentation
"""

print("Single quotes:", single_quotes)
print("Double quotes:", double_quotes)
print("Triple quotes:", triple_quotes)
```

```python
# String methods
text = "  Python Programming  "

print(f"Original: '{text}'")
print(f"Upper: '{text.upper()}'")
print(f"Lower: '{text.lower()}'")
print(f"Stripped: '{text.strip()}'")
print(f"Split: {text.strip().split()}")
print(f"Replace: '{text.replace('Python', 'Java')}'")
print(f"Join: {'-'.join(text.strip().split())}")
```

```python
# String formatting
name = "Alice"
age = 30
city = "New York"

# f-strings (recommended)
f_string = f"{name} is {age} years old and lives in {city}"
print("f-string:", f_string)

# .format() method
format_string = "{} is {} years old and lives in {}".format(name, age, city)
print(".format():", format_string)

# % operator (old style)
percent_string = "%s is %d years old and lives in %s" % (name, age, city)
print("% operator:", percent_string)
```
