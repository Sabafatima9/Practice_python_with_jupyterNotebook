## Overview
Control structures are fundamental building blocks of programming that allow you to control the flow of execution in your code. Python provides several control structures including conditional statements, loops, and exception handling.

## Types of Control Structures

### 1. Sequential Control
- Code executes line by line in order

### 2. Selection Control (Conditional)
- `if`, `elif`, `else` statements
- `match` statement (Python 3.10+)

### 3. Iteration Control (Loops)
- `for` loops
- `while` loops
- Loop control statements: `break`, `continue`, `pass`

### 4. Exception Control
- `try`, `except`, `finally`, `raise`

### 5. Function Control
- `return`, `yield`
- Function calls and recursion

## Sequential Control
The default flow where statements execute in order from top to bottom.

```python
# Sequential execution example
print("Step 1: Initialize variables")
x = 10
y = 20

print("Step 2: Perform calculations")
sum_result = x + y
product = x * y

print("Step 3: Display results")
print(f"Sum: {sum_result}")
print(f"Product: {product}")

print("Step 4: Program completed")
```

## Selection Control (Conditional Statements)

### If Statement
The basic conditional statement that executes code only if a condition is True.

```python
# Basic if statement
age = 18

if age >= 18:
    print("You are an adult")
    print("You can vote")
    print("You can drive")

print("This always executes")
```

```python
# If-else statement
score = 75

if score >= 60:
    print("Pass")
    print("Congratulations!")
else:
    print("Fail")
    print("Better luck next time")

print("Grade evaluation completed")
```

```python
# If-elif-else statement
grade = 85

if grade >= 90:
    print("Grade: A")
    print("Excellent!")
elif grade >= 80:
    print("Grade: B")
    print("Good job!")
elif grade >= 70:
    print("Grade: C")
    print("Satisfactory")
elif grade >= 60:
    print("Grade: D")
    print("Needs improvement")
else:
    print("Grade: F")
    print("Failed")

print("Grade evaluation completed")
```

```python
# Nested if statements
age = 25
income = 50000

if age >= 18:
    print("Age requirement met")
    if income >= 30000:
        print("Income requirement met")
        print("Loan approved!")
    else:
        print("Income requirement not met")
        print("Loan denied due to insufficient income")
else:
    print("Age requirement not met")
    print("Loan denied due to age")

print("Loan application processed")
```

### Match Statement (Python 3.10+)
A more powerful alternative to if-elif chains for pattern matching.

```python
# Match statement example
def analyze_command(command):
    match command.split():
        case ["help"]:
            return "Showing help information"
        case ["search", query]:
            return f"Searching for: {query}"
        case ["download", filename, "--force"]:
            return f"Force downloading: {filename}"
        case ["download", filename]:
            return f"Downloading: {filename}"
        case ["quit"] | ["exit"]:
            return "Goodbye!"
        case _:
            return "Unknown command"

# Test the match statement
commands = ["help", "search python", "download file.txt", "download file.txt --force", "quit"]

for cmd in commands:
    result = analyze_command(cmd)
    print(f"Command: '{cmd}' -> {result}")
```

## Iteration Control (Loops)

### For Loop
Used to iterate over sequences (lists, tuples, strings, etc.) or execute a block of code a specific number of times.

```python
# Basic for loop with range
print("Counting from 1 to 5:")
for i in range(1, 6):
    print(f"Number: {i}")

print("\nCounting by 2s:")
for i in range(0, 11, 2):
    print(f"Even number: {i}")

print("\nCounting backwards:")
for i in range(5, 0, -1):
    print(f"Countdown: {i}")
print("Blast off!")
```

```python
# For loop with different iterables
# List iteration
fruits = ["apple", "banana", "cherry", "date"]
print("Fruits:")
for fruit in fruits:
    print(f"- {fruit}")

# String iteration
print("\nCharacters in 'Python':")
for char in "Python":
    print(f"Character: {char}")

# Tuple iteration
coordinates = [(1, 2), (3, 4), (5, 6)]
print("\nCoordinates:")
for x, y in coordinates:
    print(f"Point: ({x}, {y})")

# Dictionary iteration
person = {"name": "Alice", "age": 30, "city": "New York"}
print("\nPerson details:")
for key, value in person.items():
    print(f"{key}: {value}")
```

```python
# Enumerate function for index and value
colors = ["red", "green", "blue", "yellow"]

print("Colors with index:")
for index, color in enumerate(colors):
    print(f"{index}: {color}")

print("\nColors with custom start index:")
for index, color in enumerate(colors, start=1):
    print(f"{index}: {color}")

# Zip function for parallel iteration
names = ["Alice", "Bob", "Charlie"]
ages = [25, 30, 35]
cities = ["New York", "London", "Paris"]

print("\nPeople information:")
for name, age, city in zip(names, ages, cities):
    print(f"{name} is {age} years old and lives in {city}")
```

### While Loop
Executes a block of code as long as a condition is True.

```python
# Basic while loop
count = 1
print("Counting with while loop:")
while count <= 5:
    print(f"Count: {count}")
    count += 1

print(f"Final count: {count}")
```

```python
# While loop with user input
print("Number guessing game (simulated)")
secret_number = 42
attempts = 0
max_attempts = 5

# Simulate user guesses
guesses = [10, 50, 30, 42, 25]
guess_index = 0

while attempts < max_attempts:
    attempts += 1
    guess = guesses[guess_index]
    guess_index += 1
    
    print(f"Attempt {attempts}: You guessed {guess}")
    
    if guess == secret_number:
        print(f"Congratulations! You found it in {attempts} attempts!")
        break
    elif guess < secret_number:
        print("Too low!")
    else:
        print("Too high!")
    
    if attempts == max_attempts:
        print(f"Game over! The number was {secret_number}")
```

```python
# While loop with continue
print("Printing even numbers from 1 to 10:")
i = 1
while i <= 10:
    if i % 2 != 0:  # If odd number
        i += 1
        continue  # Skip to next iteration
    print(f"Even number: {i}")
    i += 1

print("Loop completed")
```

## Loop Control Statements

### Break Statement
Exits the loop immediately.

```python
# Break statement example
print("Searching for number 7:")
for i in range(1, 11):
    print(f"Checking {i}")
    if i == 7:
        print(f"Found {i}! Breaking out of loop.")
        break
    print(f"{i} is not 7")

print("Search completed")
```

```python
# Break in nested loops
print("Searching in 2D grid:")
found = False

for row in range(3):
    for col in range(3):
        value = row * 3 + col + 1
        print(f"Checking position ({row}, {col}): {value}")
        
        if value == 7:
            print(f"Found 7 at position ({row}, {col})!")
            found = True
            break  # Breaks inner loop
    
    if found:
        break  # Breaks outer loop

print("Search completed")
```

### Continue Statement
Skips the current iteration and continues with the next one.

```python
# Continue statement example
print("Processing numbers (skipping multiples of 3):")
for i in range(1, 11):
    if i % 3 == 0:
        print(f"Skipping {i} (multiple of 3)")
        continue
    
    print(f"Processing {i}")
    # Simulate some processing
    result = i * 2
    print(f"  Result: {result}")

print("Processing completed")
```

## Exception Control

### Try-Except Statement
Handles exceptions (errors) that may occur during program execution.

```python
# Basic try-except
try:
    number = int(input("Enter a number: "))
    result = 100 / number
    print(f"100 / {number} = {result}")
except ValueError:
    print("Error: Please enter a valid number")
except ZeroDivisionError:
    print("Error: Cannot divide by zero")
except Exception as e:
    print(f"Unexpected error: {e}")
else:
    print("Calculation completed successfully")
finally:
    print("This always executes")

print("Program continues...")
```

```python
# Simulating the input for demonstration
def safe_division():
    test_inputs = ["10", "0", "abc", "5"]
    
    for test_input in test_inputs:
        print(f"\nTesting input: '{test_input}'")
        try:
            number = int(test_input)
            result = 100 / number
            print(f"100 / {number} = {result}")
        except ValueError:
            print("Error: Please enter a valid number")
        except ZeroDivisionError:
            print("Error: Cannot divide by zero")
        except Exception as e:
            print(f"Unexpected error: {e}")
        else:
            print("Calculation completed successfully")
        finally:
            print("This always executes")
    
    print("\nAll tests completed")

safe_division()
```

## Function Control

### Return Statement
Exits the function and returns a value.

```python
# Function with return statement
def find_maximum(numbers):
    if not numbers:
        return None  # Early return for empty list
    
    max_num = numbers[0]
    for num in numbers[1:]:
        if num > max_num:
            max_num = num
    
    return max_num

# Test the function
test_lists = [[1, 5, 3, 9, 2], [10], [], [-5, -10, -1]]

for test_list in test_lists:
    result = find_maximum(test_list)
    if result is not None:
        print(f"Maximum of {test_list}: {result}")
    else:
        print(f"Maximum of {test_list}: No elements")

# Multiple return statements
def categorize_number(num):
    if num > 0:
        return "positive"
    elif num < 0:
        return "negative"
    else:
        return "zero"

print(f"\n5 is {categorize_number(5)}")
print(f"-3 is {categorize_number(-3)}")
print(f"0 is {categorize_number(0)}")
```

## Practical Examples

### Example 1: Menu-Driven Program

```python
def calculator_menu():
    while True:
        print("\n=== Calculator Menu ===")
        print("1. Addition")
        print("2. Subtraction")
        print("3. Multiplication")
        print("4. Division")
        print("5. Exit")
        
        choice = input("Enter your choice (1-5): ")
        
        if choice == '5':
            print("Goodbye!")
            break
        elif choice in ['1', '2', '3', '4']:
            try:
                num1 = float(input("Enter first number: "))
                num2 = float(input("Enter second number: "))
                
                if choice == '1':
                    result = num1 + num2
                    operation = "+"
                elif choice == '2':
                    result = num1 - num2
                    operation = "-"
                elif choice == '3':
                    result = num1 * num2
                    operation = "*"
                elif choice == '4':
                    if num2 == 0:
                        print("Error: Cannot divide by zero!")
                        continue
                    result = num1 / num2
                    operation = "/"
                
                print(f"{num1} {operation} {num2} = {result}")
                
            except ValueError:
                print("Error: Please enter valid numbers!")
        else:
            print("Invalid choice! Please enter 1-5.")

# Simulate the menu program
print("Calculator Menu Demo (simulated)")
print("This demonstrates the control structure logic")
print("In a real program, this would run interactively")
```

### Example 2: Data Processing with Control Structures

```python
# Data processing example
def process_student_data(students):
    """Process student data and generate statistics"""
    if not students:
        print("No students to process")
        return
    
    total_score = 0
    passed = 0
    failed = 0
    excellent = 0
    
    print("Processing student data...")
    
    for student in students:
        name = student['name']
        score = student['score']
        
        total_score += score
        
        if score >= 90:
            grade = 'A'
            excellent += 1
        elif score >= 80:
            grade = 'B'
        elif score >= 70:
            grade = 'C'
        elif score >= 60:
            grade = 'D'
        else:
            grade = 'F'
        
        if score >= 60:
            passed += 1
            status = "PASSED"
        else:
            failed += 1
            status = "FAILED"
        
        print(f"{name}: {score} ({grade}) - {status}")
    
    # Calculate statistics
    average = total_score / len(students)
    
    print(f"\n=== STATISTICS ===")
    print(f"Total students: {len(students)}")
    print(f"Average score: {average:.2f}")
    print(f"Passed: {passed}")
    print(f"Failed: {failed}")
    print(f"Excellent (A): {excellent}")
    
    # Return summary
    return {
        'total': len(students),
        'average': average,
        'passed': passed,
        'failed': failed,
        'excellent': excellent
    }

# Test data
student_data = [
    {'name': 'Alice', 'score': 95},
    {'name': 'Bob', 'score': 78},
    {'name': 'Charlie', 'score': 88},
    {'name': 'Diana', 'score': 92},
    {'name': 'Eve', 'score': 55}
]

print("Student Data Processing Example")
print("=" * 40)
result = process_student_data(student_data)
```
