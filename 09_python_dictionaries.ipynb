## Overview
Dictionaries are mutable, unordered collections of key-value pairs. They provide fast lookup and are essential for data organization.

## Key Characteristics
- **Key-Value pairs**: Each element has a key and value
- **Unordered**: No guaranteed order (Python 3.7+ maintains insertion order)
- **Mutable**: Can be modified after creation
- **Keys must be hashable**: Strings, numbers, tuples (immutable)
- **Fast lookup**: O(1) average case complexity

```python
# Creating dictionaries
empty_dict = {}
person = {"name": "Alice", "age": 30, "city": "New York"}
grades = dict(John=85, Alice=92, Bob=78)

print(f"Empty: {empty_dict}")
print(f"Person: {person}")
print(f"Grades: {grades}")

# Accessing values
print(f"Name: {person['name']}")
print(f"Age: {person.get('age')}")
print(f"Phone: {person.get('phone', 'Not provided')}")
```

```python
# Dictionary operations
student = {"name": "Charlie", "age": 20}

# Adding/updating
student["grade"] = "A"
student.update({"major": "CS", "year": 2})

# Removing
removed = student.pop("year")
del student["age"]

print(f"Student: {student}")
print(f"Keys: {list(student.keys())}")
print(f"Values: {list(student.values())}")
print(f"Items: {list(student.items())}")
```

## Summary
Dictionaries are essential for:
- Data organization and lookup
- Configuration storage
- Database-like structures
- Fast data retrieval
