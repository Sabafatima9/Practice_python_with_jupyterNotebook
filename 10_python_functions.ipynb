## Overview
Functions are reusable blocks of code that perform specific tasks. They help organize code, reduce duplication, and improve maintainability.

## Key Concepts
- **Definition**: `def` keyword
- **Parameters**: Input values
- **Return**: Output values
- **Scope**: Variable visibility
- **Recursion**: Self-calling functions

```python
# Basic function
def greet(name):
    return f"Hello, {name}!"

def add(a, b=2):
    return a + b

print(greet("Alice"))
print(f"Sum: {add(5, 3)}")

# Function with default parameters
def power(base, exponent=2):
    return base ** exponent

print(f"5² = {power(5)}")
print(f"2³ = {power(2, 3)}")
```

```python
# Variable arguments
def sum_all(*args):
    return sum(args)

def person_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print(f"Sum: {sum_all(1, 2, 3, 4, 5)}")
person_info(name="Bob", age=25, city="London")
```

## Summary
Functions provide:
- Code reusability
- Modularity
- Better organization
- Easier testing and debugging
