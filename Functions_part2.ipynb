### Fibonacci

```python
def fib(n):
    if (n==0 or n==1):
       return 1
    return fib(n-1)+fib(n-2)

fib(5)
```

## Unit Converter

```python
print("Please enter a valid numeric value.")
def metre(m):
    feet = 3.2804 * m
    inch = 39.27 * m
    print(f"{m} meters = {feet} ft and {inch} inch")

def feet(f):
    metre = 0.308 * f
    inch = 12 * f
    print(f"{f} feet = {metre} meter and {inch} inch")

def inch(i):
    meter = 0.0254 * i
    feet = 0.083 * i
    print(f"{i} inches = {meter} meter and {feet} feet")
```

## Command  Line Interface(CLI)

```python
while True:
    unit = input("Welcome to Unit Converter\nChoose:\n m for meter\n f for feet\n i for inch\n q for quit\n")

    if unit == 'q' or unit == 'Q':
        break

    val = float(input("Enter a value to convert: "))

    if unit == 'm' or unit == 'M':
        metre(val)
    elif unit == 'f' or unit == 'F':
        feet(val)
    elif unit == 'i' or unit == 'I':
        inch(val)
    else:
        print("Please enter a valid character (m, f, i, q).")
```

```python
import tkinter as tk
from tkinter import ttk, messagebox

# -------------------- Functions -------------------- #
def convert():
    try:
        value = float(entry.get())
    except ValueError:
        messagebox.showerror("Error", "Please enter a valid numeric value!")
        return

    unit = unit_var.get()

    if unit == "Meter":
        feet = value * 3.28084
        inch = value * 39.3701
        result.set(f"{value:.2f} Meter\n\n= {feet:.2f} Feet\n= {inch:.2f} Inches")

    elif unit == "Feet":
        meter = value * 0.3048
        inch = value * 12
        result.set(f"{value:.2f} Feet\n\n= {meter:.2f} Meter\n= {inch:.2f} Inches")

    elif unit == "Inch":
        meter = value * 0.0254
        feet = value * 0.083333
        result.set(f"{value:.2f} Inches\n\n= {meter:.2f} Meter\n= {feet:.2f} Feet")


def clear():
    entry.delete(0, tk.END)
    result.set("Your result will appear here")


def hover_on(e):
    convert_btn["bg"] = "#1E40AF"

def hover_off(e):
    convert_btn["bg"] = "#2563EB"


# -------------------- Window -------------------- #
root = tk.Tk()
root.title("✨ Smart Unit Converter")
root.geometry("500x500")
root.config(bg="#E0F2FE")
root.resizable(False, False)

# -------------------- Card -------------------- #
frame = tk.Frame(root, bg="white", bd=0)
frame.place(relx=0.5, rely=0.5, anchor="center", width=430, height=430)

# -------------------- Heading -------------------- #
title = tk.Label(
    frame,
    text="SMART UNIT CONVERTER",
    bg="white",
    fg="#0F172A",
    font=("Segoe UI", 20, "bold")
)
title.pack(pady=20)

# -------------------- Entry -------------------- #
tk.Label(
    frame,
    text="Enter Value",
    bg="white",
    font=("Segoe UI", 11)
).pack()

entry = tk.Entry(
    frame,
    font=("Segoe UI", 14),
    justify="center",
    width=18,
    relief="solid",
    bd=1
)
entry.pack(pady=10)

# -------------------- Dropdown -------------------- #
tk.Label(
    frame,
    text="Choose Unit",
    bg="white",
    font=("Segoe UI", 11)
).pack()

unit_var = tk.StringVar()
combo = ttk.Combobox(
    frame,
    textvariable=unit_var,
    values=["Meter", "Feet", "Inch"],
    state="readonly",
    font=("Segoe UI", 11),
    width=16
)
combo.current(0)
combo.pack(pady=10)

# -------------------- Buttons -------------------- #
convert_btn = tk.Button(
    frame,
    text="Convert",
    bg="#2563EB",
    fg="white",
    font=("Segoe UI", 12, "bold"),
    width=15,
    relief="flat",
    command=convert
)
convert_btn.pack(pady=8)

convert_btn.bind("<Enter>", hover_on)
convert_btn.bind("<Leave>", hover_off)

tk.Button(
    frame,
    text="Clear",
    bg="#10B981",
    fg="white",
    font=("Segoe UI", 11, "bold"),
    width=15,
    relief="flat",
    command=clear
).pack(pady=5)

# -------------------- Result -------------------- #
result = tk.StringVar()
result.set("Your result will appear here")

result_box = tk.Label(
    frame,
    textvariable=result,
    bg="#F8FAFC",
    fg="#1E3A8A",
    font=("Segoe UI", 13, "bold"),
    relief="groove",
    width=28,
    height=6,
    bd=2,
    justify="center"
)
result_box.pack(pady=20)

# -------------------- Exit -------------------- #
tk.Button(
    frame,
    text="Exit",
    bg="#EF4444",
    fg="white",
    font=("Segoe UI", 11, "bold"),
    width=12,
    relief="flat",
    command=root.destroy
).pack()

root.mainloop()
```
