## Overview
Sets are unordered collections of unique elements in Python. They are mutable and support mathematical set operations.

## Key Characteristics
- **Unordered**: Elements have no specific order
- **Unique**: No duplicate elements allowed
- **Mutable**: Can be modified after creation
- **Hashable elements**: Elements must be immutable
- **Mathematical operations**: Union, intersection, difference

```python
# Creating sets
empty_set = set()
numbers = {1, 2, 3, 4, 5}
from_list = set([1, 2, 2, 3, 3, 4])
from_string = set("hello")

print(f"Empty: {empty_set}")
print(f"Numbers: {numbers}")
print(f"From list: {from_list}")
print(f"From string: {from_string}")
```

```python
# Set operations
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}

union = set1 | set2
intersection = set1 & set2
difference = set1 - set2
symmetric_diff = set1 ^ set2

print(f"Union: {union}")
print(f"Intersection: {intersection}")
print(f"Difference: {difference}")
print(f"Symmetric difference: {symmetric_diff}")
```

## Summary
Sets are useful for:
- Removing duplicates
- Mathematical set operations
- Membership testing
- Finding unique elements
