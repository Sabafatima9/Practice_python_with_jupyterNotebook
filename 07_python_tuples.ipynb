## Overview
Tuples are immutable, ordered collections in Python. They are similar to lists but cannot be modified after creation.

## Key Characteristics
- **Immutable**: Cannot be changed after creation
- **Ordered**: Elements maintain their order
- **Indexed**: Access elements by position
- **Heterogeneous**: Can contain different types
- **Hashable**: Can be used as dictionary keys

```python
# Creating tuples
empty_tuple = ()
single_element = (42,)
coordinates = (3, 4)
person = ("John", 30, "Engineer")

print(f"Empty: {empty_tuple}")
print(f"Single: {single_element}")
print(f"Coordinates: {coordinates}")
print(f"Person: {person}")

# Tuple unpacking
x, y = coordinates
name, age, job = person
print(f"x={x}, y={y}")
print(f"{name} is {age} years old and works as {job}")
```

```python
# Tuple operations
tuple1 = (1, 2, 3)
tuple2 = (4, 5, 6)

combined = tuple1 + tuple2
repeated = tuple1 * 3

print(f"Combined: {combined}")
print(f"Repeated: {repeated}")
print(f"Length: {len(combined)}")
print(f"Contains 5: {5 in combined}")
```

## Summary
Tuples are immutable collections useful for:
- Data that shouldn't change
- Dictionary keys
- Function return values
- Named constants
